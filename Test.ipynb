```python
import numpy as np
import matplotlib.pyplot as plt

temp = np.load('temp_magnetization.npy')
```

```python
norm  = [np.linalg.norm(temperature) for temperature in temp]

```

```python
temp_range = np.linspace(0.1, 1, 20)
plt.plot(temp_range, np.array(norm)/16)
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
%matplotlib widget


fig = plt.figure()
ax = Axes3D(fig)
for i in range(len(temp)): 
    ax.scatter(temp[i][0], temp[i][1], temp[i][2])
```
